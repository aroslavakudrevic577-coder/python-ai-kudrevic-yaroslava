```python
# Установка необходимых библиотек (если отсутствуют)
!pip install squarify plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
import plotly.express as px

# Настройка стиля
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Проверка, что df_films существует (из нулевой ячейки)
if 'df_films' not in globals():
    raise NameError("❌ df_films не найден. Сначала выполните нулевую ячейку с загрузкой данных.")

# Единая функция определения региона (используется в нескольких графиках)
def get_region(loc):
    loc_lower = loc.lower()
    if any(x in loc_lower for x in ['los angeles', 'new york', 'california', 'hollywood', 'usa', 'united states', 'american']):
        return 'США'
    elif any(x in loc_lower for x in ['london', 'uk', 'pinewood', 'british', 'england', 'united kingdom']):
        return 'Великобритания'
    elif any(x in loc_lower for x in ['paris', 'france', 'french']):
        return 'Франция'
    elif any(x in loc_lower for x in ['india', 'delhi', 'mumbai', 'bombay']):
        return 'Индия'
    elif any(x in loc_lower for x in ['australia', 'sydney', 'melbourne']):
        return 'Австралия'
    elif any(x in loc_lower for x in ['canada', 'toronto', 'vancouver']):
        return 'Канада'
    elif any(x in loc_lower for x in ['germany', 'berlin', 'munich']):
        return 'Германия'
    elif any(x in loc_lower for x in ['italy', 'rome', 'milan']):
        return 'Италия'
    elif any(x in loc_lower for x in ['spain', 'madrid', 'barcelona']):
        return 'Испания'
    elif any(x in loc_lower for x in ['japan', 'tokyo']):
        return 'Япония'
    elif any(x in loc_lower for x in ['china', 'beijing', 'shanghai']):
        return 'Китай'
    elif any(x in loc_lower for x in ['mexico', 'mexico city']):
        return 'Мексика'
    elif any(x in loc_lower for x in ['brazil', 'são paulo', 'rio']):
        return 'Бразилия'
    elif any(x in loc_lower for x in ['europe', 'amsterdam', 'netherlands', 'sweden', 'norway', 'denmark', 'finland', 'poland', 'czech', 'austria', 'switzerland', 'belgium']):
        return 'Другие страны Европы'
    else:
        return 'Прочие'

# Цветовая карта регионов (глобальная)
region_colors = {
    'США': '#1f77b4',
    'Великобритания': '#ff7f0e',
    'Франция': '#2ca02c',
    'Индия': '#d62728',
    'Австралия': '#9467bd',
    'Канада': '#8c564b',
    'Германия': '#e377c2',
    'Италия': '#7f7f7f',
    'Испания': '#bcbd22',
    'Япония': '#17becf',
    'Китай': '#aec7e8',
    'Мексика': '#ffbb78',
    'Бразилия': '#98df8a',
    'Другие страны Европы': '#c5b0d5',
    'Прочие': '#c49c94'
}

# ============================================================
# График 1: Гистограмма распределения кассовых сборов
# ============================================================
print("\n💡 Инсайт 1: Большинство фильмов имеют сборы до 500 млн долларов, а фильмы со сборами выше 1 млрд встречаются крайне редко.")
plt.figure(figsize=(14, 6))
plt.hist(df_films['box_office'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.yscale('log')
plt.xlabel('Кассовые сборы (USD)', fontsize=12)
plt.ylabel('Частота (log scale)', fontsize=12)
plt.title('Распределение кассовых сборов фильмов (логарифмическая шкала)', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ============================================================
# График 2: Топ-20 локаций по количеству фильмов
# ============================================================
print("\n💡 Инсайт 2: Лос-Анджелес, Нью-Йорк и Калифорния — безусловные лидеры по числу съёмок, что отражает концентрацию киноиндустрии в США.")
top_locations_count = df_films['location'].value_counts().head(20).sort_values()
plt.figure(figsize=(14, 8))
top_locations_count.plot(kind='barh', color='coral', edgecolor='black')
plt.xlabel('Количество фильмов', fontsize=12)
plt.ylabel('Локация', fontsize=12)
plt.title('Топ-20 локаций по числу снятых фильмов', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)
for i, v in enumerate(top_locations_count):
    plt.text(v + 5, i, str(v), va='center', fontsize=10)
plt.tight_layout()
plt.show()

# ============================================================
# График 3 (улучшенный): Топ-20 локаций по суммарным кассовым сборам
# ============================================================
print("\n💡 Инсайт 3: Дели (один фильм «Жил-был тигр») генерирует огромные сборы, обгоняя многие популярные локации.")

# Подготовка данных
location_revenue = df_films.groupby('location')['box_office'].sum().sort_values(ascending=False).head(20).sort_values()
df_plot = location_revenue.reset_index()
df_plot.columns = ['location', 'total_revenue']
df_plot['region'] = df_plot['location'].apply(get_region)

print("Распределение регионов в топ-20:")
print(df_plot['region'].value_counts())

colors = df_plot['region'].map(region_colors).tolist()

# Построение графика
plt.figure(figsize=(14, 8))
bars = plt.barh(df_plot['location'], df_plot['total_revenue'], color=colors, edgecolor='black')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Явно задаём тики оси X
xticks = ax.get_xticks()
ax.set_xticks(xticks)
ax.set_xticklabels([f'{int(x/1e9)}' for x in xticks])
plt.xlabel('Суммарные кассовые сборы (млрд USD)', fontsize=12)
plt.ylabel('Локация', fontsize=12)
plt.title('Топ-20 локаций по суммарным кассовым сборам', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Числа справа от столбцов
for i, (_, row) in enumerate(df_plot.iterrows()):
    plt.text(row['total_revenue'] + 0.3e9, i, f'{row["total_revenue"]/1e9:.1f}',
             va='center', fontsize=10, color='black')

# Легенда
unique_regions = df_plot['region'].unique()
handles = [plt.Rectangle((0,0),1,1, color=region_colors[reg]) for reg in unique_regions]
plt.legend(handles, unique_regions, title='Регион', loc='lower right', fontsize=9)

plt.tight_layout()
plt.show()

# ============================================================
# Комбинированный график: Топ-20 локаций – сборы (столбцы) и количество фильмов (линия)
# ============================================================
print("\n💡 Инсайт: Сравнение суммарных сборов и количества фильмов позволяет оценить эффективность локаций. Например, Дели имеет максимальные сборы при минимуме фильмов, а Лос-Анджелес – много фильмов, но сборы на фильм ниже.")

# Подготовка данных
loc_stats = df_films.groupby('location').agg(
    total_revenue=('box_office', 'sum'),
    film_count=('film', 'count')
).reset_index()
loc_stats = loc_stats.sort_values('total_revenue', ascending=False).head(20).copy()
loc_stats['region'] = loc_stats['location'].apply(get_region)
colors = loc_stats['region'].map(region_colors).tolist()

# Фигура
fig, ax1 = plt.subplots(figsize=(16, 9))

# Столбцы
x = range(len(loc_stats))
bars = ax1.bar(x, loc_stats['total_revenue'], color=colors, edgecolor='black', alpha=0.8)

ax1.set_xlabel('Локация', fontsize=12)
ax1.set_ylabel('Суммарные кассовые сборы (млрд USD)', fontsize=12)
ax1.set_title('Топ-20 локаций по сборам: сравнение сборов и количества фильмов', fontsize=14, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(loc_stats['location'], rotation=45, ha='right', fontsize=10)
ax1.grid(axis='y', linestyle='--', alpha=0.5)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Форматирование оси Y
yticks = ax1.get_yticks()
ax1.set_yticks(yticks)
ax1.set_yticklabels([f'{int(y/1e9)}' for y in yticks])

# Подписи на столбцах
for i, (_, row) in enumerate(loc_stats.iterrows()):
    ax1.text(i, row['total_revenue'] + 0.2e9, f'{row["total_revenue"]/1e9:.1f}',
             ha='center', va='bottom', fontsize=8, color='black')

# Правая ось для количества фильмов
ax2 = ax1.twinx()
ax2.plot(x, loc_stats['film_count'], color='red', marker='o', linestyle='-', linewidth=2, markersize=8)
ax2.set_ylabel('Количество фильмов', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.spines['right'].set_visible(True)
ax2.spines['top'].set_visible(False)

# Подписи для линии (справа от точек)
for i, (_, row) in enumerate(loc_stats.iterrows()):
    ax2.text(i + 0.15, row['film_count'], str(int(row['film_count'])),
             ha='left', va='center', fontsize=9, color='red')

# Легенда
unique_regions = loc_stats['region'].unique()
handles_bars = [plt.Rectangle((0,0),1,1, color=region_colors[reg]) for reg in unique_regions]
handles_line = [plt.Line2D([0], [0], color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='Количество фильмов')]
handles = handles_bars + handles_line
labels = list(unique_regions) + ['Количество фильмов']

ax1.legend(handles, labels, title='Регион / Показатель', loc='upper left',
           bbox_to_anchor=(1.02, 1), fontsize=9, borderaxespad=0)

# Вместо tight_layout используем subplots_adjust, чтобы легенда не обрезалась
plt.subplots_adjust(right=0.85)
plt.show()

# ============================================================
# График 4: Связь популярности и средних сборов (scatter plot)
# ============================================================
print("\n💡 Инсайт 4: Локации с небольшим числом фильмов могут иметь очень высокие средние сборы за счёт отдельных кассовых хитов, в то время как популярные локации показывают средние результаты.")
loc_stats = df_films.groupby('location').agg(
    film_count=('film', 'count'),
    total_revenue=('box_office', 'sum'),
    avg_revenue=('box_office', 'mean')
).reset_index()
loc_stats_filtered = loc_stats[loc_stats['film_count'] >= 5]

plt.figure(figsize=(14, 8))
sc = plt.scatter(
    loc_stats_filtered['film_count'],
    loc_stats_filtered['avg_revenue'] / 1e6,
    s=loc_stats_filtered['total_revenue'] / 1e7,
    c=loc_stats_filtered['total_revenue'] / 1e9,
    alpha=0.6,
    cmap='viridis',
    edgecolors='black',
    linewidth=0.5
)
plt.colorbar(sc, label='Общие сборы (млрд USD)')
plt.xlabel('Количество фильмов в локации', fontsize=12)
plt.ylabel('Средние сборы (млн USD)', fontsize=12)
plt.title('Зависимость средних сборов от популярности локации', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ============================================================
# График 5: Ящик с усами для топ-10 локаций (логарифмическая шкала)
# ============================================================
print("\n💡 Инсайт 5: В Лос-Анджелесе и Нью-Йорке наблюдается огромный разброс сборов: от малобюджетных фильмов до блокбастеров. В студийных локациях (Pinewood) сборы в среднем выше.")
top10_locs = df_films['location'].value_counts().head(10).index
df_top10 = df_films[df_films['location'].isin(top10_locs)]

plt.figure(figsize=(16, 8))
sns.boxplot(
    data=df_top10,
    x='location',
    y='box_office',
    hue='location',          # добавляем hue для устранения предупреждения
    palette='Set2',
    order=top10_locs,
    legend=False             # легенда не нужна, так как ось X уже подписана
)
plt.yscale('log')
plt.xlabel('Локация', fontsize=12)
plt.ylabel('Кассовые сборы (USD, log scale)', fontsize=12)
plt.title('Распределение кассовых сборов в топ-10 локациях по числу фильмов', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ============================================================
# График 6: Treemap долей суммарных сборов по локациям (интерактивный, Plotly)
# ============================================================
print("\n💡 Инсайт 6: Несколько локаций (Лос-Анджелес, Нью-Йорк, Дели) аккумулируют основную долю кассовых сборов. Остальные распределены равномерно.")

# Группируем по локациям
loc_sum = df_films.groupby('location')['box_office'].sum().reset_index()
loc_sum = loc_sum.sort_values('box_office', ascending=False).reset_index(drop=True)
loc_sum = loc_sum[loc_sum['box_office'] > 0]

# Берём топ-150 локаций, остальные объединяем
top_n = 150
if len(loc_sum) <= top_n:
    treemap_data = loc_sum.copy()
else:
    top_locs = loc_sum.iloc[:top_n].copy()
    other_sum = loc_sum.iloc[top_n:]['box_office'].sum()
    other_row = pd.DataFrame({
        'location': [f'Другие ({len(loc_sum)-top_n} локаций)'],
        'box_office': [other_sum]
    })
    treemap_data = pd.concat([top_locs, other_row], ignore_index=True)

# Строим интерактивный treemap с помощью Plotly
fig = px.treemap(
    treemap_data,
    path=[px.Constant("Все локации"), 'location'],
    values='box_office',
    title='Treemap: доли локаций в суммарных кассовых сборах',
    color='box_office',
    color_continuous_scale='Viridis',
    hover_data={'box_office': ':,.0f'}
)

fig.update_layout(
    width=1200,
    height=900,
    font=dict(size=14),
    title_font_size=20
)

fig.update_traces(
    textinfo="label+percent entry",
    textfont_size=14,
    hovertemplate='<b>%{label}</b><br>Сборы: $%{value:,.0f}<extra></extra>'
)

fig.show()

# ============================================================
# График 7: Топ-10 фильмов по кассовым сборам с локациями (исправленный)
# ============================================================
print("\n💡 Инсайт 7: Фильмы-гиганты часто снимаются в нескольких локациях одновременно, что отражает глобальный масштаб производства (например, «Аватар» — Калифорния, Новая Зеландия, Гавайи).")

# Подготовка данных
film_locations = df_films.groupby('film')['location'].apply(lambda x: ', '.join(x)).reset_index()
film_revenue = df_films.groupby('film')['box_office'].first().reset_index()
film_top = pd.merge(film_revenue, film_locations, on='film')
film_top = film_top.sort_values('box_office', ascending=False).head(10)

# Увеличиваем размер фигуры и настраиваем отступы вручную
plt.figure(figsize=(16, 10))
bars = plt.barh(film_top['film'], film_top['box_office'] / 1e6, color='gold', edgecolor='black')

# Настройка осей и заголовка
plt.xlabel('Кассовые сборы (млн USD)', fontsize=12)
plt.ylabel('Фильм', fontsize=12)
plt.title('Топ-10 фильмов по кассовым сборам', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()

# Добавляем подписи локаций справа от столбцов
for i, (_, row) in enumerate(film_top.iterrows()):
    plt.text(row['box_office']/1e6 + 15, i, row['location'],
             va='center', fontsize=9, color='darkblue')

# Сетка
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Ручная настройка отступов, чтобы избежать предупреждения tight_layout
plt.subplots_adjust(left=0.25, right=0.85, top=0.95, bottom=0.1)

plt.show()

print("\n✅ Все 7 графиков успешно построены!")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Функция определения региона
def get_region(location):
    # Список стран Европы (можно расширить при необходимости)
    europe = ['UK', 'United Kingdom', 'England', 'France', 'Germany', 'Italy', 'Spain',
              'Великобритания', 'Англия', 'Франция', 'Германия', 'Италия', 'Испания',
              'London', 'Лондон', 'Paris', 'Париж', 'Pinewood Studios']
    # Проверка: если локация содержит название европейской страны или города
    loc_lower = str(location).lower()
    for e in europe:
        if e.lower() in loc_lower:
            return 'Europe'
    # США
    if 'usa' in loc_lower or 'united states' in loc_lower or 'california' in loc_lower or 'new york' in loc_lower or 'los angeles' in loc_lower:
        return 'USA'
    return 'Other'

region_colors = {
    'Europe': '#2ca02c',   # зелёный
    'USA': '#1f77b4',      # синий
    'Other': '#ff7f0e'     # оранжевый
}

# ============================================================
# Комбинированный график: Топ-20 локаций – сборы (столбцы) и количество фильмов (линия)
# ============================================================
print("\n💡 Инсайт: Сравнение суммарных сборов и количества фильмов позволяет оценить эффективность локаций. Например, Дели имеет максимальные сборы при минимуме фильмов, а Лос-Анджелес – много фильмов, но сборы на фильм ниже.")

# Подготовка данных
loc_stats = df_films.groupby('location').agg(
    total_revenue=('box_office', 'sum'),
    film_count=('film', 'count')
).reset_index()
loc_stats = loc_stats.sort_values('total_revenue', ascending=False).head(20).copy()
loc_stats['region'] = loc_stats['location'].apply(get_region)
colors = loc_stats['region'].map(region_colors).tolist()

# Фигура
fig, ax1 = plt.subplots(figsize=(16, 9))

# Столбцы
x = range(len(loc_stats))
bars = ax1.bar(x, loc_stats['total_revenue'], color=colors, edgecolor='black', alpha=0.8)

ax1.set_xlabel('Локация', fontsize=12)
ax1.set_ylabel('Суммарные кассовые сборы (млрд USD)', fontsize=12)
ax1.set_title('Топ-20 локаций по сборам: сравнение сборов и количества фильмов', fontsize=14, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(loc_stats['location'], rotation=45, ha='right', fontsize=10)
ax1.grid(axis='y', linestyle='--', alpha=0.5)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Форматирование оси Y
yticks = ax1.get_yticks()
ax1.set_yticks(yticks)
ax1.set_yticklabels([f'{int(y/1e9)}' for y in yticks])

# Подписи на столбцах (чёрные цифры) с ручной корректировкой для некоторых городов
manual_adjust_revenue = {
    'Paris': 0.5e9,       # поднимаем выше на 0.5 млрд
    'New Jersey': 0.4e9,
    'Arizona': 0.4e9,
    # можно добавить другие
}
for i, (_, row) in enumerate(loc_stats.iterrows()):
    y_offset = manual_adjust_revenue.get(row['location'], 0.2e9)  # по умолчанию +0.2 млрд
    ax1.text(i, row['total_revenue'] + y_offset, f'{row["total_revenue"]/1e9:.1f}',
             ha='center', va='bottom', fontsize=8, color='black')

# Правая ось для количества фильмов
ax2 = ax1.twinx()
ax2.plot(x, loc_stats['film_count'], color='red', marker='o', linestyle='-', linewidth=2, markersize=8)
ax2.set_ylabel('Количество фильмов', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.spines['right'].set_visible(True)
ax2.spines['top'].set_visible(False)

# Подписи для линии (красные цифры) с ручной корректировкой
manual_adjust_film = {
    # для этих городов поднимаем цифру выше точки
    'Paris': ('above', 2),     # поднять на 2 единицы
    'New Jersey': ('above', 2),
    'Arizona': ('above', 2),
    # для городов, где точка внутри столбца, опускаем цифру под точку
    'Los Angeles': ('below', -1.5),
    'New York': ('below', -1.5),
    'California': ('below', -1.5),
    # можно добавить другие
}
for i, (_, row) in enumerate(loc_stats.iterrows()):
    loc = row['location']
    y_val = row['film_count']
    if loc in manual_adjust_film:
        pos, delta = manual_adjust_film[loc]
        if pos == 'above':
            y_text = y_val + delta
            va = 'bottom'
        else:
            y_text = y_val + delta
            va = 'top'
    else:
        y_text = y_val
        va = 'bottom' if y_val < max(loc_stats['film_count']) * 0.7 else 'top'
    ax2.text(i + 0.15, y_text, str(int(row['film_count'])),
             ha='left', va=va, fontsize=9, color='red')

# Легенда внутри графика справа вверху с полупрозрачным фоном
unique_regions = loc_stats['region'].unique()
handles_bars = [plt.Rectangle((0,0),1,1, color=region_colors[reg]) for reg in unique_regions]
handles_line = [plt.Line2D([0], [0], color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='Количество фильмов')]
handles = handles_bars + handles_line
labels = list(unique_regions) + ['Количество фильмов']

ax1.legend(handles, labels, title='Регион / Показатель',
           loc='upper right', fontsize=9,
           facecolor='white', edgecolor='gray', framealpha=0.8)

# Убираем лишний отступ справа, так как легенда теперь внутри
plt.subplots_adjust(right=0.95)
plt.show()
```
